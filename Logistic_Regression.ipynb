#Logistic Regression Model A:

1. initialization: set learning rate, iterations, weights, and bias
2.  training (fit):
 * compute predictions using sigmoid
 * calculate gradients
 * update weights and bias using gradient descent
3.  prediction (predict):
 * apply learned weights to new data
 * use sigmoid to classify the data into 0 or 1

```python
import math;
#includes gradient descent for optimization

class LogisticRegression: #class definition encapsulates logistic reg model
    def _init_(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

#   initializes key parameters:
#     learning_rate: controls how much the model updates the weights in each iteration.
#     iterations: number of times gradient descent runs to optimize the weights
#     weights: stores the coefficients for the features (initialized later)
#     bias: stores the bias term (initialized later)


    def sigmoid(self, z):
        return 1 / (1 + math.exp(-z))
#   implements sigmoid function; maps any input value z to prob 0 or 1


    def fit(self, X, y):
        m, n = len(X), len(X[0])  # m= number of samples, n= number of features per sample
        self.weights = [0] * n    # initializes weights to zero (a list of size n).
        self.bias = 0             # initializes bias to zero.

        for _ in range(self.iterations): # runs gradient descent for iterations number of items
            dw = [0] * n                 # dw: gradient (change needed) for each weight
            db = 0                       # db: gradient for the bias

            for i in range(m):
                linear_model = sum(self.weights[j] * X[i][j] for j in range(n)) + self.bias
                y_predicted = self.sigmoid(linear_model)

            #  calculates the linear model
            #  z= w1x1 + w2x2 + ... + wnxn + b
            #  applies the sigmoid function to compute the predicted probability= y_predicted

                error = y_predicted - y[i]     # calculates the error between predicted and actual values
                for j in range(n):
                    dw[j] += error * X[i][j] / m        # dw[j] accumulates the gradient for each weight
                db += error / m                         # db accumulates the gradient for the bias

            for j in range(n):
                self.weights[j] -= self.learning_rate * dw[j]
            self.bias -= self.learning_rate * db

            # updates each weight using the gradient and learning rate
            # updates the bias in a similar way

    def predict(self, X):
        predictions = []              # takes in new input X to classify
        for i in range(len(X)):
            linear_model = sum(self.weights[j] * X[i][j] for j in range(len(X[0]))) + self.bias
            y_predicted = self.sigmoid(linear_model)
            # computes the linear model for each sample and applies the sigmoid function

            predictions.append(1 if y_predicted >= 0.5 else 0)
            # if probability is >= 0.5, classify as 1, else 0
        return predictions
```

#Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#given data
X= np.array ([1,2,3])
Y= np.array ([2,2.8,3.6])
```

```python
# Initialize parameters
beta_0=0
beta_1=0
alpha=0.1 #learning rate (alpha updates the beta by this much )
epochs=500 #number of iterations
```

```python
m=len(X) #number of data points
```

```python
#Gradient Descent
for epoch in range(epochs):
    y_pred=beta_0+beta_1*X #compute predictions
    error=y_pred-Y #compute error

    #compute gradients
    grad_beta_0=(2/m)*np.sum(error)
    grad_beta_1=(2/m)*np.sum(error*X)

    #compute parameters
    beta_0 -=alpha*grad_beta_0
    beta_1 -=alpha*grad_beta_1

    #update parameters
    beta_0=beta_0-alpha*grad_beta_0
    beta_1=beta_1-alpha*grad_beta_1

    print(f"Iteration {epoch+1}:beta_0={beta_0:.3f}, beta_1 = {beta_1:.3f}")

    print(f"\nFinal equation: h(x)={beta_0:.3f}+{beta_1:.3f}x")
```

```python
#Simulated training and validation loss over epochs
epochs = np.arange(1, 201)
train_loss = np.exp(-epochs/50) + 0.02*np.random.rand(len(epochs))
val_loss = np.exp(-epochs/50) + 0.02*np.random.rand(len(epochs)) + 0.05*np.sin(epochs/10)

```

```python
#Plotting
plt.figure(figsize=(8,6))
plt.plot(epochs, train_loss, label="Training Loss", color="blue")
plt.plot(epochs, val_loss, label="Validation Loss", color="red")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs. Validation Loss")
plt.axvline(x=75, linestyle="--", color="gray", label="Best Epoch")
plt.legend()
plt.show()
```

#Gradient Descent with Early Stopping
Step 1: Generate Sample Data
We’ll create a dataset where Y = 3X + 5 + noise and train a model to estimate w (weight) and b (bias).

Step 2: Implement Gradient Descent with Early Stopping
We’ll track validation loss and stop training if it doesn’t improve after a few iterations.



```python
#Generate Data (Y= 3X + 5 + noise)
np.random.seed(42)
X=np.linspace(0, 10, 100) #Returns num evenly spaced samples, calculated over the interval [start, stop]
Y=3*X+5+np.random.randn(100)*2 #adding noise
```

```python
X
```

```python
#Train-test split
split= int(0.8*len(X))
X_train, X_val= X[:split],X[split:]
Y_train, Y_val= Y[:split],Y[split:]
```

```python
#initialize parameters
w=np.random.randn() # Weight
b=np.random.randn() # Bias

alpha= 0.01 #Learning rate
epochs= 1000 #max epochs
patience= 20 #number of epochs to wait before stopping if no improvement
best_loss= float('inf') #track best validation loss
wait = 0 #counter for patience
```

```python
train_losses=[]
val_losses=[]

#training loop with early stopping
for epoch in range(epochs):
    #forward pass: predictions
    Y_pred= w * X_train + b
    val_pred=w * X_val + b

    #compute loss (Mean squared error)
    train_loss= np.mean((Y_train-Y_pred)**2)
    val_loss= np.mean((Y_val-val_pred)**2)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    #backward pass: compute gradients
#    dw= -2*np.mean((Y_train-Y_pred)*X_train)
#    db= -2*np.mean(Y_train-Y_pred)

    #early stopping check
    if val_loss < best_loss:
      best_loss= val_loss
      wait=0 # reset patience counter
    else:
      wait +=1 #increase counter
      if wait >= patience:
        print(f"Early stopping at epoch {epoch} (Best Val Loss: {best_loss:.4f})")
        break

    #Compute Gradients (slopes)
    d_w= (2/len(X_train))*np.sum((Y_pred-Y_train)*X_train) #len(X_train) = m (number of variables)
    d_b= (2/len(X_train))*np.sum(Y_pred-Y_train)

    #update parameters
    w -= alpha * d_w
    b -= alpha * d_b

    #plot loss curve
    plt.plot(train_losses, label="Train Loss", color="blue")
    plt.plot(val_losses, label="Val Loss", color="red")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training vs. Validation Loss (Early Stopping)")
    plt.legend()
    plt.show()

    print(f"Final Weight: {w:.4f}, Final Bias: {b:.4f}")
```

# Building the Model from Scratch

```python
#importing the dependencies
import numpy as np #mathematical and scientific computing library; like matrix multiplication
import pandas as pd
```

**Logistic regression**

def init: initiating the parameters

def fit: fit dataset to our logistic regression model; fit the X input features to predict the outcome Y

def update weights: using gradient descent; weight and bias are initially randomly assigned. gradient descent model will update the weight/bias

def predict: will tell us if prediction is 1 or 0

===================================

**Work flow of the Logistic Regression model:**

Step 1: Set the Learning Rate & number of iterations; Initiate random weight and bias value

Step 2: Build Logistic Regression function (sigmoid function)

Step 3: Update the parameters using gradient descent

Finally, we will get the best model (best weight and bias value) as it has minimum cost function

Step 4: build the "predict" function to determine the class of the data point

```python
class Logistic_Regression(): #provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state; car - blueprint. blueprint=class; car=object


  #declaring learning rate and number of iterations (hyperparameters)
  def __init__(self, learning_rate, no_of_iterations):
    self.learning_rate= learning_rate
    self.no_of_iterations= no_of_iterations



  #fit function to train the model with dataset
  def fit(self, X, Y):
    #number of data points * number of input features

    #determine number of data points (rows) in data set and number of features (variables/columns)
    #consider it a matrix 8 inputs and 769 data points: [769 x 8]
    # n -> number of input features
    # m -> number of points
    # m and n are the parameters of this model, we represent them within class LR, must write as self.m... SELF makes a reference to the model itself

    # number of data points in the dataset (number of rows) --> m
    # number of input features in the dataset (number of columns) --> n
        # n -  important to determine how many weights we will have
    self.m, self.n = X.shape #defines matrix [m,n]

    # initiating weight and bias value
    self.w = np.zeros(self.n) #creates an array of n weights, ex n=8: [0. 0. 0. 0. 0. 0. 0. 0.]
    self.b = 0
    self.X = X #represents input dataset
    self.Y = Y #represents output column

    # create gradient descent algorithm to go through the data a thousand times (iterations) and update the values to minimize the cost function
    # model = logistic regression of learning rate and number of iterations
    # to update weight we need these 4 formulas:
      # 1) w2(updated weight)=w1+a*dw
      # 2) b2(updated bias)=b1+a*dw
    # learning rate
      # 3) dw = (1/m)* (Y^-Y).X
      # 4) db = (1/m)* (Y^-Y)


    #cost function goes data point by data point
    # doing the loss > sums all of the errors and we update to try to minimize
    # Y^ (predicted outcome) = sigmoid equation ... ranges from 0 to 1
      # Y^ .. what is the probability that your y value is = 1 ?
      # ex Y^ = 0.8 > 80% chance that your Y value = 1

    #THIS IS WHERE GRADIENT DESCENT STARTS
    # implementing gradient descent for optimization

    for i in range(self.no_of_iterations):
      self.update_weights()




  def update_weights(self):

    #Y^ formula (sigmoid function); np.exp(1) = 2.718281828459045 ... e^100000 = large number.. e^-23782340 ~= 0.
    # with sigmoid func: with + Z = 1/1 = Y ~1 ;
                      # with - Z = 1/(1+e^-(-34980943...)) 1/large number = Y ~= 0

    # Z=w.X+b
    Y_hat= 1/(1+np.exp(-(self.X.dot(self.w+self.b)))) #takes dot product; matrix multiplication


    # derivatives
    dw= (1/self.m)*np.dot(self.X.T,(Y_hat - self.Y))

    db= (1/self.m)*np.sum(Y_hat - self.Y)

    # updating weights & bias using gradient descent
      # gradient descent -> upd_w=w-a*dw ; upd_b=b-a*db
      # FINDS UPDATED WEIGHT AND BIAS until iterations end
    self.w = self.w - self.learning_rate*dw

    self.b = self.b - self.learning_rate*db

  # Sigmoid Equation & Decision Boundary/ threshold
  def predict(self, X):

    Y_pred = 1/(1+np.exp(-(X.dot(self.w+self.b)))) #Y_hat equation
    Y_pred = np.where(Y_pred > 0.5, 1, 0) #if value of Y_pred is greater> 0.5 = 1 otherwise, 0
    return Y_pred

```
